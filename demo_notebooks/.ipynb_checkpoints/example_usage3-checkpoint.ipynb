# User guide notebook - Pipeline from Autogluon models

<span style="color:orange">NOTE 1 : </span> It is crucial to stratify your data while performing the train test split, so that there are no classes in test data that were not present in the train set. <br>
<span style="color:orange">NOTE 2 : </span> It is crucial to convert binary target column labels into 0 for negative class and 1 for positive class. Otherwise, some evaluation metrics may not be calculated correctly.

```python
import pandas as pd
from pathlib import Path
import sys
from autogluon.tabular import TabularPredictor
from sklearn.model_selection import train_test_split
BASE_DIR = Path.cwd()
import os
project_root = BASE_DIR.parent
src_path = project_root/"src"
sys.path.append(str(src_path))

#make sure the Project root directory is correct [...]/Rashomon_analysis
print(BASE_DIR.parent)

from rashomon_analysis.pipelines.builder_abstract import *
from rashomon_analysis.pipelines.pipelines_user_input import *
```

## Build Rashomon Set pipeline from Autogluon input

## Provide all necessary arguments:
all you need to initizalize the pipeline is TabularPredictor object, dataset as TabularDataset objects

```python
#load Autogluon predictor from your local directory
path_to_autogluon_arguments = BASE_DIR/'demo_models'/'autogluon'/'credit'
predictor = TabularPredictor.load(path_to_autogluon_arguments)

#load dataset the models were trained on
data = pd.read_csv(project_root/"demo_data"/"credit_score.csv")
target_column = 'Credit Score'

#split dataset into train and test - only test data will be used in the pipeline
#stratify to make sure there are no classes not present in test set
train, test = train_test_split(data, test_size=0.2, random_state=42, stratify=data[target_column])

#ensure indices are reset to get valid sample indices later on
train = train.reset_index(drop=True)
test = test.reset_index(drop=True)
#convert test data to TabularDataset object
test_data = TabularDataset(test)

print(type(predictor))
print(type(test_data))
```

## Build pipeline

```python
builder = BuildRashomonAutogluon(predictor = predictor, 
                                test_data = test,
                                df_name = 'credit',
                                base_metric='accuracy',
                                delta = 0.05,
                                feature_imp_needed = True,
                                converter_results_directory = BASE_DIR/'demo_models'/'autogluon'/'credit'/'converter_results')
```

## Preview Rashomon Set properties

```python
builder.preview_rashomon()
```

### Set epsilon based on the preview plot

```python
builder.set_epsilon(0.031)
```

### Launch pipeline  and visualize the dashboard

```python
rashomon_set, visualizer = builder.build()
```

### Work on the Rashomon Set object 

```python
print(rashomon_set.base_model)
```

```python
rashomon_set.summarize_rashomon()
```

### Remember to close streamlit dashboard after you are finished

```python
builder.dashboard_close()
```
